```python
import os
current_folder = os.getcwd()

novels_folder_path = os.path.join(current_folder, '..\\..\\insight_text_game_dataset\\visual-novel-preprocessing-byQMark-v2')
novels_folderName_list = os.listdir(novels_folder_path)

save_folder_path = os.path.join(current_folder, '..\\..\\insight_text_game_dataset\\visual-novel-traningData')
if not os.path.exists(save_folder_path):
    os.makedirs(save_folder_path, exist_ok=True)
```

```python
def data_split(json_data):
    novel = []
    input_novel=[]
    answer = []
    output_novel=[]
    result=[]
    
    q=0
    a=0
    
    for line in json_data:
        str_value = list(line.keys())
        
        if("question" in str_value):
            q=1
            if(input_novel and answer and output_novel):
                result.append({"input_novel":input_novel})
                result.append({"answer":answer})
                result.append({"output_novel":output_novel})
                novel.append(result.copy())
                
                input_novel = output_novel.copy()
                
                answer.clear()
                output_novel.clear()
                result.clear()
                a=0
                
            continue
        if("answer" in str_value):
            a=1
            continue
                
        if q==0 and a==0:
            input_novel.append(line)
        elif a==0 and q==1:
            answer.append(line)
        elif q==1 and a==1:
            output_novel.append(line)
            
    return novel
```

```python
import json

def save_novel(preprocessing_novel_json, novel_folder_name, novel_name):
    save_novelfolder_path = os.path.join(save_folder_path, novel_folder_name)
    
    def save_json():
        save_json_url = os.path.join(save_novelfolder_path, novel_name)
        write_novel_json(save_json_url)
        
    def create_save_folder_of_novel():
        if not os.path.exists(save_novelfolder_path):
            os.makedirs(save_novelfolder_path, exist_ok=True)
    
    def write_novel_json(save_json_url):
        with open(save_json_url, 'wt', encoding='utf-8') as f:
            json.dump(preprocessing_novel_json, f, ensure_ascii=False, indent=4)
            
    create_save_folder_of_novel()
    save_json()
```

```python
import json

for novel_folder_name in novels_folderName_list[1:]:
    novel_list = os.listdir(os.path.join(novels_folder_path, novel_folder_name))
    for novel_name in novel_list:
        novel_path = os.path.join(novels_folder_path, novel_folder_name, novel_name)
        with open(novel_path, 'r', encoding='utf-8') as f:
            json_data = json.load(f)
            preprocessing_novel_json = data_split(json_data)
            save_novel(preprocessing_novel_json, novel_folder_name, novel_name)
```
