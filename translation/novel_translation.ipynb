```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast


model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")

```

```python
import os

current_path = os.getcwd()
dataset_path = os.path.join(current_path,"..\\dataset")
    
novelSet_path = os.path.join(dataset_path, "visual-novels")
save_novel_path = os.path.join(dataset_path, "visual-novels-ko")

print("novelSet_path: ", novelSet_path)

novel_list = os.listdir(novelSet_path)
```

```python
def create_folder(_path):
    if not os.path.exists(_path):
        os.makedirs(_path, exist_ok=True)

create_folder(dataset_path)
create_folder(save_novel_path)
```

```python
def translation(novel_title, novel_content):
    cnt=0
    
    with open(novel_title, 'w', encoding='utf-8') as f:
        for i in novel_content:
            cnt+=1
            tokenizer.src_lang = "en_XX"
            encoded_hi = tokenizer(i, return_tensors="pt")
            generated_tokens = model.generate(
                **encoded_hi,
                forced_bos_token_id=tokenizer.lang_code_to_id["ko_KR"]
            )
            result = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
            f.write(str(result) + '\n')  # 각 문장을 줄바꿈과 함께 저장
            if cnt%25==0:
                print(f"translated {cnt} sentences")
```

```python
for novel_name in novel_list:
    open_file = os.path.join(novelSet_path, novel_name)
    f = open(open_file, 'r', encoding='utf-8')
    
    save_path_ko = os.path.join(save_novel_path, novel_name)
    cnt=0
    
    while True:
        line = f.readline()
        if not line: 
            break
        
        with open(save_path_ko, 'w', encoding='utf-8') as fr:
            cnt+=1 
            tokenizer.src_lang = "en_XX"
            encoded_hi = tokenizer(line, return_tensors="pt")
            generated_tokens = model.generate(
                **encoded_hi,
                forced_bos_token_id=tokenizer.lang_code_to_id["ko_KR"]
            )
            result = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
            fr.write(str(result) + '\n')  # 각 문장을 줄바꿈과 함께 저장
            if cnt%25==0:
                print(f"translated {cnt} sentences")
            print(result)
    f.close()
    print(f'complete novel : {novel_name}')
```
